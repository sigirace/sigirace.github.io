**1. ÏÉÅÏûê**

- Îç∞Ïù¥ÌÑ∞ Í∞íÏùò 50%Í∞Ä Ìï¥ÎãπÌï®
- ÏÉÅÏûêÏùò ÏïÑÎû´Î©¥Í≥º ÏúóÎ©¥ÏùÄ Í∞Å 25%(1ÏÇ¨Î∂ÑÏúÑÏàò), 75%(3ÏÇ¨Î∂ÑÏúÑÏàò) ÏúÑÏπòÏóê Ï°¥Ïû¨ÌïòÎäî Í∞í
- ÏÉÅÏûê Ï§ëÏïôÏùò ÎëêÍ∫ºÏö¥ ÏÑ†ÏùÄ Ï§ëÏïôÍ∞íÏùÑ ÏùòÎØ∏Ìï®
- Ï§ëÏïôÍ∞íÏùò ÏúÑÏπòÏóê Îî∞Îùº Í∞íÏù¥ Ïñ¥ÎîîÏóê ÏπòÏö∞Ï≥êÏ†∏ ÏûàÎäîÏßÄ ÌôïÏù∏Ìï† Ïàò ÏûàÏùå

**2. IQR**

- Ï§ëÏïô 50%Îç∞Ïù¥ÌÑ∞Í∞Ä ÌçºÏßÑ Ï†ïÎèÑ
- 3ÏÇ¨Î∂ÑÏúÑÏàòÏôÄ 1ÏÇ¨Î∂ÑÏúÑÏàòÏùò Ï∞®Ïù¥Î°ú Íµ¨Ìï† Ïàò ÏûàÏùå
- ÌÜµÍ≥ÑÌïôÏ†ÅÏúºÎ°ú Ïù¥ÏÉÅÏπòÎ•º ÌåêÎã®ÌïòÎäî ÏàòÏãùÏóê ÏÇ¨Ïö©Îê®

**3. ÏàòÏóº**

- 1.5 * IQR Î≤îÏúÑÏù∏ Q1 - 1.5 * IQR Î∂ÄÌÑ∞ Q3 + 1.5 * IQR ÍπåÏßÄÏùò Î≤îÏúÑÎ•º ÏàòÏóºÏúºÎ°ú ÏßÄÏ†ï (1.5Í∞Ä ÏïÑÎãå Îã§Î•∏ Í∞íÏúºÎ°úÎèÑ ÏÇ¨Ïö© Í∞ÄÎä•)
- ÏàòÏóºÏùÑ Î≤óÏñ¥ÎÇòÎäî Í∞íÏùÄ Ïù¥ÏÉÅÏπòÎ°ú ÌåêÎã®

üßë‚Äçüíª : Box plotÏùÄ ÏàòÏπòÌòï Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ÏÑú Ï†ÅÏö©ÌïòÍ∏∞Ïóê ÏàòÏπòÌòï Ïª¨ÎüºÎì§ÏùÑ Î®ºÏ†Ä Ï†ïÏùòÌïòÍ≤†ÏäµÎãàÎã§.

üí¨ **Chat Prompt**

> Ï†úÏãúÌïú Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ïª¨ÎüºÏùÑ ÏàòÏπòÌòï, Î≤îÏ£ºÌòïÏúºÎ°ú Íµ¨Î∂ÑÌïòÏó¨ ÏïåÎ†§Ï£ºÍ≥† Î¶¨Ïä§Ìä∏Ïóê Ï†ÄÏû•Ìï¥Ï§ò

```python
# ÏàòÏπòÌòï Ïª¨ÎüºÍ≥º Î≤îÏ£ºÌòï Ïª¨ÎüºÏúºÎ°ú Íµ¨Î∂Ñ
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns.drop('ID').tolist()
categorical_columns = data.select_dtypes(exclude=['float64', 'int64']).columns.tolist()

# Í≤∞Í≥º Ï∂úÎ†•
numeric_columns, categorical_columns
```

‚ù£Ô∏è **ÍøÄÌåÅ**

- Îç∞Ïù¥ÌÑ∞Ïùò Ïä§ÏºÄÏùºÏù¥ Îã§Î•¥Îã§Î©¥ ÌïòÎÇòÏùò ÌôîÎ©¥Ïóê Î™®Îì† Ïª¨ÎüºÏóê ÎåÄÌïú ÏãúÍ∞ÅÌôîÎ≥¥Îã§Îäî Í∞úÎ≥ÑÏ†ÅÏù∏ ÏãúÍ∞ÅÌôîÍ∞Ä ÌååÏïÖÏù¥ Ïâ¨ÏõÄ
- Í∞úÎ≥ÑÏ†ÅÏúºÎ°ú Í∑∏Î¶¨Îêò ÌïòÎÇòÏùò ÌôîÎ©¥Ïóê Í∑∏Î¶¥ Ïàò ÏûàÎäî Í∑∏Î¶¨ÎìúÎ•º ÌôúÏö©

üí¨ **Chat Prompt**
> ÏàòÏπòÌòï Ïª¨ÎüºÎì§Ïóê ÎåÄÌï¥ÏÑú Í∞úÎ≥ÑÏ†ÅÏúºÎ°ú box plot ÏãúÍ∞ÅÌôî ÏàòÌñâ<br>
> Îã®, box plotÏùÄ Í∞úÎ≥ÑÏ†ÅÏúºÎ°ú ÏàòÌñâÌïò Îêò, ÌïòÎÇòÏùò Í∑∏Î¶¨ÎìúÏóê ÌëúÌòÑ

```python
# ÏàòÏπòÌòï Ïª¨ÎüºÎì§Ïóê ÎåÄÌï¥ Box plotÏùÑ ÌïòÎÇòÏùò Í∑∏Î¶¨ÎìúÏóê ÌëúÌòÑ
plt.figure(figsize=(18, 12))

for i, column in enumerate(numeric_columns, 1):
    plt.subplot(4, 5, i)  # 4x5 Í∑∏Î¶¨Îìú ÏÑ§Ï†ï
    plt.boxplot(data[column].dropna())
    plt.title(f'{column}')
    plt.xticks([])

plt.tight_layout()
plt.show()
```

üí¨ **Chat Prompt**

> IQR Í¥ÄÏ†êÏóêÏÑú Ïù¥ÏÉÅÏπòÍ∞Ä ÏûàÎäî Ïª¨ÎüºÎì§ÏùÑ ÏïåÎ†§Ï§ò

ü§ñ **GPT Ï∂úÎ†•: Message**

‚úèÔ∏è Î∂ôÏó¨ÎÑ£Í∏∞

ü§ñ **GPT Ï∂úÎ†•: Code**

```python
# IQRÏùÑ Í≥ÑÏÇ∞ÌïòÏó¨ Ïù¥ÏÉÅÏπòÍ∞Ä ÏûàÎäî Ïª¨ÎüºÏùÑ Ï∞æÎäî Ìï®Ïàò
def detect_outliers_iqr(df, columns):
    outliers = []
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        if df[(df[column] < lower_bound) | (df[column] > upper_bound)].shape[0] > 0:
            outliers.append(column)
    return outliers

# Ïù¥ÏÉÅÏπòÍ∞Ä ÏûàÎäî Ïª¨Îüº ÌôïÏù∏
outlier_columns = detect_outliers_iqr(data, numeric_columns)
outlier_columns
```

- ÌôïÏó∞Ïù¥ Î≤îÏúÑÎ•º Î≤óÏñ¥ÎÇòÎäî Ïù¥ÏÉÅÏπòÍ∞Ä ÏïÑÎãå Í≤ΩÏö∞ ÎèÑÎ©îÏù∏ Ï†ÑÎ¨∏Í∞ÄÏôÄ Ï≤òÎ¶¨ Î∞©ÏïàÏóê ÎåÄÌïú ÌòëÏùòÍ∞Ä ÌïÑÏöîÌï®
- ÌäπÌûà Ïó∞ÏÜçÏ†ÅÏù∏ Ïù¥ÏÉÅÏπòÍ∞Ä ÎÇòÌÉÄÎÇòÎäî Í≤ΩÏö∞Îäî Ïù¥ÏÉÅÏπòÍ∞Ä ÏïÑÎãê ÌôïÎ•†Ïù¥ ÎÜíÏùå
- ÎèÑÎ©îÏù∏ Ï†ÑÎ¨∏Í∞ÄÏùò ÎèÑÏõÄÏùÑ Î∞õÏßÄ Î™ªÌïòÎäî ÏÉÅÌô©ÏóêÏÑúÎäî Î≥¥ÏàòÏ†ÅÏúºÎ°ú Ïù¥ÏÉÅÏπòÎ•º Ï†úÍ±∞ (ex, ÏÉÅÏúÑ 5%, Í∞ÄÏû• ÌÅ∞ Í∞í, ...)
- IQR ÎßåÏúºÎ°ú Ï†úÍ±∞ÌïúÎã§Î©¥ ÎÑàÎ¨¥ ÎßéÏùÄ Îç∞Ïù¥ÌÑ∞Î•º Ï†úÍ±∞Ìï† Ïàò ÏûàÏùå

## 5. ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ

ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎûÄ 2Í∞úÏùò Î≥ÄÏàò Í∞ÑÏùò ÏÑ†ÌòïÍ¥ÄÍ≥ÑÎ•º ÌëúÌòÑÌïòÎäî ÌÜµÍ≥ÑÏ†Å Ï∏°ÎèÑÏûÖÎãàÎã§. 3Î≤à Ìï≠Î™©ÏóêÏÑú ÏÇ∞Ï†êÎèÑÎ•º ÌÜµÌï¥ Î≥ÄÏàò ÏÇ¨Ïù¥Ïùò Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨Î•º ÏÇ¥Ìé¥Î≥¥Í≥† Í¥ÄÍ≥ÑÏùò Ïú†ÌòïÍ≥º Í∞ïÎèÑÎ•º ÌåêÎã®Ìï¥ Î≥¥ÏïòÏäµÎãàÎã§. Ïù¥Î≤à ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÏãúÍ∞ÅÌôîÏóêÏÑúÎäî Ïù¥Î•º ÏàòÏπòÌôîÌïòÏó¨ Îç∞Ïù¥ÌÑ∞ Í∞Ñ Í¥ÄÍ≥ÑÎ•º ÌëúÌòÑÌï©ÎãàÎã§.

üìç **ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Ìï¥ÏÑùÌïòÍ∏∞**

- ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎäî -1~1 ÏÇ¨Ïù¥Ïùò Ïà´Ïûê Í∞íÏúºÎ°ú Ï∂úÎ†•Îê® ‚òû ÌîºÏñ¥Ïä® ÏÉÅÍ¥ÄÍ≥ÑÏàò
- ÌîºÏñ¥Ïä® ÏÉÅÍ¥ÄÍ≥ÑÏàòÎäî ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º ÏàòÏπòÌôîÌïòÎäî ÌÜµÍ≥ÑÏ†Å Í∏∞Î≤ï
- ÌîºÏñ¥Ïä® ÏÉÅÍ¥ÄÍ≥ÑÏàòÎäî Îç∞Ïù¥ÌÑ∞Ïùò Ïä§ÏºÄÏùºÍ≥º Î¨¥Í¥ÄÌïòÍ∏∞Ïóê Ï†ÑÏ≤òÎ¶¨ Ï†ÑÏóê ÏàòÌñâÌï¥ÎèÑ Î¨¥Î∞©
- ÏïÑÎûò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌåêÎã® Í∏∞Ï§Ä ÌëúÏóê Îî∞Îùº Ìï¥ÏÑù

| Î≤îÏúÑ             | ÏÑ§Î™Ö               | Î≤îÏúÑ           | ÏÑ§Î™Ö               |
| ---------------- | ------------------ | -------------- | ------------------ |
| r <= -0.8        | Í∞ïÌïú ÏùåÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ | 0.8 <= r       | Í∞ïÌïú ÏñëÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ |
| -0.8 < r <= -0.6 | ÏùåÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ      | 0.6 <= r < 0.8 | ÏñëÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ      |
| -0.6 < r <= -0.4 | ÏïΩÌïú ÏùåÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ | 0.4 <= r < 0.6 | ÏïΩÌïú ÏñëÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ |
| -0.4 < r <= 0    | Í±∞Ïùò ÏÉÅÍ¥Ä ÏóÜÏùå     | 0 <= r < 0.4   | Í±∞Ïùò ÏÉÅÍ¥Ä ÏóÜÏùå     |



üí¨ **Chat Prompt**
> ÏàòÏπòÌòï Î≥ÄÏàòÎì§Ïóê ÎåÄÌï¥ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌñâÎ†¨ÏùÑ ÎßåÎì§Í≥† Ïù¥Î•º ÏãúÍ∞ÅÌôîÌï¥Ï§ò

ü§ñ **GPT Ï∂úÎ†•: Message**

‚úèÔ∏è Î∂ôÏó¨ÎÑ£Í∏∞

ü§ñ **GPT Ï∂úÎ†•: Code**

```python
import seaborn as sns

# ÏàòÏπòÌòï Î≥ÄÏàòÎì§Ïóê ÎåÄÌï¥ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌñâÎ†¨ ÏÉùÏÑ±
correlation_matrix = data[numeric_columns].corr()

# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌñâÎ†¨ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('ÏàòÏπòÌòï Î≥ÄÏàòÎì§ Í∞ÑÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌñâÎ†¨')
plt.show()
```

üßë‚Äçüíª: Î™©Ï†ÅÏù¥ ÎêòÎäî Ï¢ÖÏÜçÎ≥ÄÏàòÏôÄ Ïù¥Î•º ÏÑ§Î™ÖÌïòÍ∏∞ ÏúÑÌïú ÎèÖÎ¶ΩÎ≥ÄÏàòÍ∞ÑÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÎÜíÏùÄ Í≤ÉÏùÄ Îß§Ïö∞ Ï¢ãÏùÄ ÌòïÏÉÅÏûÖÎãàÎã§. ÌïòÏßÄÎßå ÎèÖÎ¶ΩÎ≥ÄÏàòÎÅºÎ¶¨ Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÏûàÎã§Î©¥ Îã§Ï§ëÍ≥µÏÑ†ÏÑ±ÏùÑ ÏùòÏã¨Ìï¥Ïïº Ìï©ÎãàÎã§.

üìç **Îã§Ï§ëÍ≥µÏÑ†ÏÑ±**

ÌöåÍ∑ÄÎ∂ÑÏÑùÏóêÏÑú ÏÑ§Î™ÖÎ≥ÄÏàòÍ∞ÑÏùò Í∞ïÌïú ÏÑ†ÌòïÍ¥ÄÍ≥ÑÍ∞Ä ÏûàÏùÑ Îïå Î∞úÏÉùÌïòÎäî Î¨∏Ï†úÎ°ú ÏÑ§Î™ÖÎ≥ÄÏàòÎÅºÎ¶¨ ÏÉÅÌò∏ÏûëÏö©ÌïòÍ±∞ÎÇò Ï§ëÎ≥µÏ†ÅÏúºÎ°ú ÏòÅÌñ•ÏùÑ ÎØ∏ÏπòÎäî ÏÉÅÌô©ÏùÑ ÎßêÌï©ÎãàÎã§. Îã§Ï§ëÍ≥µÏÑ†ÏÑ± Î¨∏Ï†úÍ∞Ä ÏûàÏùÑ Ïãú Í≤∞Í≥ºÌï¥ÏÑù Î∞è ÌöåÍ∑ÄÍ≥ÑÏàò Ï∂îÏ†ïÏù¥ Î∂àÏïàÏ†ïÌï¥ÏßÄÎäî Î¨∏Ï†úÎ•º ÏïºÍ∏∞Ìï©ÎãàÎã§.

üí¨ **Chat Prompt**
> ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌñâÎ†¨ÏùÑ ÌÜµÌï¥ Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÏûàÎäî Î≥ÄÏàòÏåçÏùÑ ÏïåÎ†§Ï§ò

ü§ñ **GPT Ï∂úÎ†•: Message**

‚úèÔ∏è Î∂ôÏó¨ÎÑ£Í∏∞

ü§ñ **GPT Ï∂úÎ†•: Code**

```python
# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌñâÎ†¨ÏóêÏÑú Ï†àÎåÄÍ∞íÏù¥ 0.7 Ïù¥ÏÉÅÏù∏ Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Í∞ÄÏßÑ Î≥ÄÏàòÏåçÏùÑ Ï∞æÍ∏∞
strong_correlation_pairs = correlation_matrix.unstack().reset_index()
strong_correlation_pairs.columns = ['Variable1', 'Variable2', 'Correlation']
strong_correlation_pairs['abs_Correlation'] = strong_correlation_pairs['Correlation'].abs()
strong_correlation_pairs = strong_correlation_pairs[strong_correlation_pairs['abs_Correlation'] >= 0.7]

# Ï§ëÎ≥µ Ï†úÍ±∞
strong_correlation_pairs = strong_correlation_pairs[strong_correlation_pairs['Variable1'] != strong_correlation_pairs['Variable2']]
strong_correlation_pairs = strong_correlation_pairs.drop_duplicates(subset=['abs_Correlation'])

print(strong_correlation_pairs)
```
